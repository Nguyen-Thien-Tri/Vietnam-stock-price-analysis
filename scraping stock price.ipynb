```python
import requests
import pandas as pd

# Function to scrape data from the API for a specific stock code
def scrape_data(stock_code):
    all_data = []
    page = 1

    while True:
        url = f"https://s.cafef.vn/Ajax/PageNew/DataHistory/PriceHistory.ashx?Symbol={stock_code}&StartDate=&EndDate=&PageIndex={page}&PageSize=20"
        response = requests.get(url)

        if response.status_code == 200:
            data = response.json()

            # Check if data exists in the response
            if 'Data' in data and 'Data' in data['Data'] and data['Data']['Data']:
                for record in data['Data']['Data']:
                    record['Company code'] = stock_code  # Add 'Company code' to each record
                    all_data.append(record)
                print(f"Successfully retrieved data for {stock_code} on page {page}")
                page += 1
            else:
                print(f"No more data found for {stock_code} on page {page}. Stopping.")
                break
        else:
            print(f"Failed to retrieve data for {stock_code} on page {page}")
            break

    return all_data

# Function to save the scraped data into an Excel file
def save_to_excel(data, filename='stock_data.xlsx'):
    # Convert the list of dictionaries to a DataFrame
    df = pd.DataFrame(data)
    # Save the DataFrame to an Excel file
    df.to_excel(filename, index=False)
    print(f"Data saved to '{filename}'")


# Read the Excel file to get the stock codes
stock_codes_df = pd.read_excel('/content/drive/MyDrive/Chứng khoán/VN30.xlsx')  # Change the file name if needed
stock_codes = stock_codes_df['Mã CK'].tolist()

# Initialize an empty list to store data for all stock codes
all_stock_data = []
```

```python
# Scrape data for each stock code and append it to the all_stock_data list
for stock_code in stock_codes:
    stock_data = scrape_data(stock_code)
    all_stock_data.extend(stock_data)

# Save the merged data into a single Excel file
save_to_excel(all_stock_data, filename='VN30_stock_data.xlsx')
```
